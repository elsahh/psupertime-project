```python
import pandas as pd
df = pd.read_csv("sample.csv", sep = ",")
df
```

```python
df["Age"] = df["Title"].apply(lambda x: x.split("yr")[0])
df["Age"] = df["Age"].astype(int)
df
```

```python
import os
p = './GSE81547_data/'
cells = pd.DataFrame()

for filename in os.listdir(p):
    f = os.path.join(p,filename)
    if os.path.isfile(f):
        cellname = f.split("_")[1].split("/")[1]
        cell = pd.read_csv(f, sep = "\t", header = None)
        cell = cell.pivot_table(columns = 0, values = 1)
        cell.rename(index={1:cellname}, inplace=True)
        cells = pd.concat([cells,cell])
cells
```

```python
cells.to_csv("GSE81547.csv", sep = ",", index = True)
```

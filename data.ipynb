```python
import pandas as pd
df = pd.read_csv("GSE81547.csv", sep = ",", index_col = 0)
df
```

```python
#drop unwanted columns
lowerCols = []
for c in df.columns:
    if c[0].islower(): lowerCols.append(c)
lowerCols

df = df[[c for c in df.columns if c not in lowerCols]]
df
```

```python
#pseudocounts and log transform
import numpy as np

logdf = df.apply(lambda x: x.apply(lambda y: y if y != 0 else 1))
logdf = np.log10(logdf)

# row-wise normalization: sum of each row adds up to one
logdf = logdf.apply(lambda x: x / sum(x), axis = 1)
logdf
```

```python
# calculate variance of each column, remove every gene below a variance threshold (insert NaN), create new data frame with remaining genes (400-500)
import numpy as np
threshold = 0.55e-07

highvargenes = logdf.apply(lambda x: x if x.var() >= threshold else [np.NaN]*len(x), axis = 0)
highvargenes.dropna(axis=1, inplace = True)
highvargenes
```

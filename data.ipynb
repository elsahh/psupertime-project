```python
import pandas as pd
df = pd.read_csv("GSE81547.csv", sep = ",", index_col = 0)
df
```

```python
#drop unwanted columns
lowerCols = []
for c in df.columns:
    if c[0].islower(): lowerCols.append(c)
lowerCols

df = df[[c for c in df.columns if c not in lowerCols]]
df
```

```python
import numpy as np

sparsedf = df.replace(0, np.nan)
sum = sparsedf.isna().sum(axis = 0)
drop = sum >= 2519
datlist = list(drop[drop].index.values)
sparsedf.drop(datlist, axis =1, inplace=True)
sparsedf.fillna(value = 0, inplace=True)
sparsedf
```

```python
#pseudocounts of 1 are added 0's to do log transformation
logdf = sparsedf.apply(lambda x: x.apply(lambda y: y if y != 0 else 1))
logdf = np.log10(logdf)

# row-wise normalization: sum of each row adds up to one
logdf = logdf.apply(lambda x: x / x.sum(), axis = 1)
logdf
```

```python
# calculate variance of each column, remove every gene below a variance threshold (insert NaN), create new data frame with remaining genes (400-500)
threshold = 0.57e-07

_500genes = logdf.apply(lambda x: x if x.var() >= threshold else [np.NaN]*len(x), axis = 0)
_500genes.dropna(axis=1, inplace = True)
_500genes
```

```python
threshold = 1.3e-07

_50genes = logdf.apply(lambda x: x if x.var() >= threshold else [np.NaN]*len(x), axis = 0)
_50genes.dropna(axis=1, inplace = True)
_50genes
```

```python
_500genes.to_csv("500_variable_genes.csv", sep = ",", index = True)
_50genes.to_csv("50_variable_genes.csv", sep = ",", index = True)
```
